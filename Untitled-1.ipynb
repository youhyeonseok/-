```python
import pandas as pd
import tqdm
import datetime
```

```python
data = pd.read_csv('서울특별시 광진구_주정차단속현황_20220610.csv',encoding='cp949')
data.head()
```

```python
data2 = pd.read_csv("서울특별시 광진구_공영주차장 정보_20220118 (1).csv", encoding='cp949')
data2.head()
```

```python
place_list = ["중곡", "구의", "광장", "화양", "능동", "군자", "자양"]
# 동마다 분석할 필요가 있음
# 그러면 동 이름을 넣으면 시각화까지 할 수 있는 모듈이 필요하다
# data, data2 에서 동 검색 검색 기능(띄어쓰기 예외처리, 군자로... 포함, 나머지는 위치를 알 수 없으므로 버림)
# 그러면 동명을 검색한 data를 가지고 일자별로 단속 횟수 다시 데이터 세팅
# data2에 운영 개시한 일자를 기준으로 앞뒤 term일 평균의 차이를 비교 
```

```python
def str_to_datetime(x):
    format = "%Y.%m.%d %H:%M"
    return datetime.datetime.strptime(x, format)

def searching_data(name, data):

    mask = [False for i in range(len(data))]
    for i in tqdm.tqdm(range(len(data))):
        curr_name = data.단속동.iloc[i].replace(' ','')
        if curr_name.find(name) != -1:
            mask[i] = True

    return data[mask].reset_index(drop = True)

def counter(scaled_data):
    yy = scaled_data.단속일시.iloc[0].year
    mm = scaled_data.단속일시.iloc[0].month
    dd = scaled_data.단속일시.iloc[0].day
    col = ["datetime","count"]
    df = pd.DataFrame(columns = col)
    count = 0
    for i in tqdm.tqdm(range(len(scaled_data))):
        curr_date = scaled_data.단속일시.iloc[i]
        if curr_date.month != mm or curr_date.day != dd:

            date = str(yy) + "-" + str(mm).zfill(2) + "-" + str(dd).zfill(2)
            line = pd.DataFrame([[date,count]],columns = col)
            df = pd.concat([df, line],axis = 0)

            yy = curr_date.year
            mm = curr_date.month
            dd = curr_date.day
            count = 0

        else:
            count += 1

    return df.set_index("datetime")
```

```python
scaled_data = searching_data("구의", data)
scaled_data.단속일시 = scaled_data.단속일시.apply(str_to_datetime)
scaled_data = counter(scaled_data)
```
